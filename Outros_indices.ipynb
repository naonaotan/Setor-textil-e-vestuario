```python
df = pd.read_csv('dfs_consolidadas.csv')
# df = df[['DENOM_CIA', 'SETOR_ATIV']]
```

```python
# df[df['DENOM_CIA'].str.contains('RENNER', case=False)].head(1)
```

```python
# selecionando o setor
df_setor = df.loc[df['SETOR_ATIV'] == 'Têxtil e Vestuário']
```

```python
# selecionando o periodo
df_ultimo = df_setor.loc[df['ORDEM_EXERC'] == 'ÚLTIMO']
```

```python
df = df_ultimo
```

```python
# df[df['DENOM_CIA'].str.contains('RENNER', case=False)]
```

```python
# filtrando os valores que precisamos da coluna CD_CONTA
df_indicadores = df[(df['CD_CONTA'] == '3.11') |
 (df['CD_CONTA'] == '2.03') |
  (df['CD_CONTA'] == '3.01') |
   (df['CD_CONTA'] == '1') |
    (df['CD_CONTA'] == '3.05')]
```

```python
# selecionando apenas as colunas que vamos usar
df_colunas = df_indicadores[['DENOM_CIA','CD_CONTA','DT_FIM_EXERC','VL_CONTA']]
```

```python
# df_colunas.head()
```

```python
# pivotando tabela
# index o que eu quero que seja linha
# columns o que eu quero que seja coluna
# values os valores
df_pivot = df_colunas.pivot_table(index=['DENOM_CIA', 'DT_FIM_EXERC'], columns='CD_CONTA', values='VL_CONTA').reset_index()
df_pivot
```

```python
# criando novas colunas baseada em calculos de colunas ja existentes
# df_pivot['nova_coluna'] = df_pivot['coluna_x'] / df_pivot['coluna_y']
df_pivot['ROA'] = df_pivot['3.11'] / df_pivot['2.03']
df_pivot['ROE'] = df_pivot['3.11'] / df_pivot['1']
df_pivot['ML'] = df_pivot['3.11'] / df_pivot['3.01']
```

```python
# df_pivot.head(1)
```

```python
df = df_pivot.rename(columns={'3.11': 'LL', '3.05': 'EBIT'})
```

```python
# df = df.reset_index()
```

```python
df = df[['DENOM_CIA', 'DT_FIM_EXERC', 'ROA', 'ROE', 'ML', 'LL', 'EBIT']]
```
